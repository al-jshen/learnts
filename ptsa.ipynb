A **stochastic process** is a family of random variables indexed with some numbers (indicating a time marker). A **time series** can be thought of as an instance or a realization of a stochastic process. Just as random variables are no longer random once an observation is made, time series are no longer random as soon as they are realized. 

Linear regression uses an intercept and a slope to predict one variable given another as an input. That is, a model might be expressed as $y = m*x+b$, where $m$ is the slope, and $b$ is the intercept. 

```r
# fitting a linear model to the co2 dataset.
co2.model.linear = lm(co2~time(co2)) 

# plot the co2 data and the fit of the linear model on top. 
plot(co2)
abline(co2.model.linear, col='blue')
```

The difference between the actual values and the predictions we get from our linear model is called the **residual**. We can analyze the normality of residuals visually using a histogram and a Q-Q plot (which plots two setse of quantiles against each other). 

```r
# taking the "residuals" (errors between the actual and our predictions) and plotting them in a histogram.
# overlaying a density on top. 
co2.model.linear.resid = resid(co2.model.linear)
hist(co2.model.linear.resid, freq=0, breaks=30)
lines(density(co2.model.linear.resid))

# checking whether the errors are normally distributed. if they are, they should follow the line without any significant systematic deviation. 
# in this case, they do NOT appear to be normal. 
qqnorm(co2.model.linear.resid)
qqline(co2.model.linear.resid)
```

**Autocovariance** is the covariance between two points in a time series. **Autocorrelation** is to autocovariance as correlation is to variance. Note that given a time lag of 0, autocorrelation is always 1 (because a variable is perfectly correlated with itself). 

A **weakly stationary process** is one where the mean is fixed over time, and the autocovariance (variance between two timesteps, say $Z_t$ and $Z_{t+\tau}$) depends only on the time lag $\tau$ (interval between timesteps), not on the actual position of the point (i.e. $Cov(Z_t, Z_{t+\tau}) = Cov(Z_{t+1}, Z_{t+\tau+1})$). In the following notes I will use "weakly stationary" and "stationary" interchangeably. 

Intuitively, stationary processes are ones which show no trend. You could examine some subsection of data, and it would be "indistinguishable" (in terms of the mean and variance) from another subsection. 

A **random walk** is a process where each timestep is the sum of the previous timestep and some noise. $$Z_i \sim N(\mu, \sigma^2)$$ $$X_0 = 0$$ $$X_1 = Z_1$$ $$X_2 = X_1 + Z_2 = Z_1 + Z_2$$ $$X_t = \sum_{i=1}^t Z_i$$ Note that the first timestep is entirely noise, and each timestep is really just the sum of all noises up to that time. 

```r
# random walk

rwalkvec = NULL
rwalkvec[1] = 0
for (i in 2:1000) {
  rwalkvec[i] = rwalkvec[i-1] + rnorm(1)
}

randomwalk = ts(rwalkvec)

plot(randomwalk)
```

Random walks are not stationary processes. 

Proof: 

$$E(X_t) = E(\sum_{i=1}^t Z_i) = \sum_{i=1}^tE(Z_i) = \sum_{i=1}^t \mu = t\mu$$

which shows that the mean changes over time ($E(X_t)$ is a function of $t$). 

---

A **moving average process of order $q$** defines an output as a linear combination of the previous $q$ outputs. For example, we might have
$$\text{for } Z_i \stackrel{i.i.d.}{\sim}N(0, \sigma^2) \\ X_t = \sum_{i=0}^q \beta_i Z_{t-i} = \beta_0Z_t + \beta_1Z_{t-1} + \ldots + \beta_t + Z_{t-q}\,{\rm, \beta_i \in R}.$$

A moving average is a stationary process.

Proof: 
$$E(X_t) = E(\sum_{i=0}^q \beta_i Z_{t-i}) = \sum_{i=0}^q \beta_i E(Z_{t-i}) = \sum_{i=0}^q \beta_i * 0 = 0$$

Now we want to examine whether $Cov(X_t, X_{t+\tau})$ is independent of $t$, i.e. $$Cov(X_t, X_{t+\tau}) = Cov(\sum_{i=0}^q\beta_iZ_{t-i}, \sum_{j=0}^{q}\beta_jZ_{t+\tau-j})$$ In particular, this is a function of some combination of $Z_i's, Z_j's$ for some $i's, j's$. Note that given some $Z_i, Z_j$, by independence, $Cov(Z_i, Z_j) = 0$ if $i \neq j$, and $Cov(Z_i, Z_j) = Var(Z_i) = Var(Z_j)$ if $i = j$. So, only the terms where $i=j$ get picked out, so this is really just a combination of all the variances $Var(Z_i)$. In this case, the terms where $i=j-\tau$ get picked out. 

In the case of $\tau=0$, and we have $q+1$ terms where $i$ and $j$ agree ($\{0, 1, \ldots, q\}$), so $$Cov(X_t, X_{t+\tau}) = \sum_{i=0}^q Var(\beta_iZ_{t-i}) = \sum_{i=0}^q \beta_i^2 Var(Z_{t-i}) = \sum_{i=0}^q \beta_i^2 \sigma^2 = \sigma^2 \sum_{i=0}^q \beta_i^2 $$
In particular, note that this is independent of $t$.

For $\tau=1$, there aren't $q+1$ terms where $i=j-\tau$ anymore. Rather, there are only $q$ terms ($\{1, 2, ..., q\}$). Furthermore, the $\beta$ coefficient will be shifted in one of the terms, since $Z_i = Z_{j-1}$ when $j=i+1$. So, the summation is becomes $$\sigma^2 \sum_{i=0}^{q-1}\beta_i\beta_{i+1}$$ 

In general, for some $\tau < q$, there are $q-\tau$ terms ($\{\tau, \tau+1, \ldots, q\}$), and the $\beta$ coefficient is shifted by $\tau$. So, the summation becomes $$\sigma^2 \sum_{i=0}^{q-\tau}\beta_i\beta_{i+\tau}$$ 

Again, this is independent of $t$, so this is a stationary process. 


```r
# moving average process, order 2

noise = rnorm(1002)

ma2 = NULL

for (i in 3:1002) {
  ma2[i] = noise[i] + 0.7 * noise[i-1] + 0.2 * noise[i-2]
}

mavg = ma2[3:1002]
mavg = ts(mavg)

plot(mavg)
```

We can generate a **correlogram** by plotting the autocorrelation for different time lags. For a moving average process of order 2, we expect there to be a non-trivial correlation only for time lags of 0, 1, and 2. 

```r
acf(mavg)
```

We can apply **transformations** to turn non-stationary processes into stationary ones. The purpose of doing so is that we have powerful techniques for analyzing stationary ones. 

One such transformation is **differencing**, where we take the difference between success points. Note that applying differencing on a set of $n$ points yields a set of $n-1$ points. We can see how differencing transforms the random walk from before into a stationary process.

```r
par(mfrow=c(2, 2))

plot(randomwalk)
acf(randomwalk)

plot(diff(randomwalk))
acf(diff(randomwalk))
```

An autoregressive process is a stochastic process that depends linearly on its own previous values, as well as a stochastic term. The output $X_t$ is defined as 

$$X_t = Z_t + \text{history} = Z_t + \phi_1 X_{t-1} + \ldots + \phi_p X_{t-p}$$
where $Z_t$ is the stochastic term (e.g. $Z_t \stackrel{i.i.d.}{\sim}N(0, \sigma^2)$). Autoregressive processes are not necessarily stationary.

```r
# AR(1) process
N <- 500
phi = 0.4
z = rnorm(N, 0, 1)
x = NULL
x[1] = z[1]
for (t in 2:N) {
    x[t] = z[t] + phi * x[t-1]
}
x.ts = ts(x)
autoplot(x.ts)
acf(x.ts)
```

```r
# AR(2) process
x.ts <- arima.sim(list(ar=c(0.6, 0.2)), n=500)
autoplot(x.ts)
acf(x.ts)
```

The **backward shift operator** is an operator $B$ defined as:

$$BX_t = X_{t-1} \\ B^2X_t = BBX_t = BX_{t-1} = X_{t-2} \\ B^kX_t = X_{t-k}$$

We can rewrite an MA(q) process as follows:

$$ X_t = \mu + \beta_0Z_t + \ldots + \beta_qZ_{t-q} \\ = \mu + \beta_0Z_t + \ldots + \beta_qB^qZ_t \\ = \mu + Z_t (\beta_0 + \ldots + \beta_q B^q)$$

We can also rewrite an AR(p) process as follows:

$$X_t = Z_t + \phi_1X_{t-1} + \ldots + \phi_pX_{t-p} \\ = Z_t + \phi_1 BX_t + \ldots + \phi_p B^pX_t \\ = Z_t + X_t (\phi_1B + \ldots + \phi_p B^p)$$

$$\implies Z_t = X_t(1-\phi_1B + \ldots + \phi_p B^p)$$

We can express an AR(1) process as an MA($\infty$) process. The reason for doing so is that it makes the (mean, variance, and autocorrelation of the) AR process easy to analyze. 

We start with the definition of an AR(1) process:

$$X_t = Z_t + \phi BX_t$$
where $Z_t \stackrel{i.i.d.}{\sim}(0, \sigma^2)$.

We rearrange to get
$$(1-\phi B)X_t = Z_t$$

By the geometric series expansion, we get
$$X_t = \frac{1}{1-\phi B}Z_t = (1+\phi B + \phi^2 B^2 + \ldots) Z_t \\ = Z_t + \phi Z_{t-1} + \phi^2 Z_{t-2} + \ldots \\ = Z_t + \alpha_1 Z_{t-1} + \alpha_2 Z_{t-2} + \ldots$$

where the coefficients $\alpha_i = \phi^i$. This is the form of an MA($\infty$) process. 

Then, we can easily compute $E(X_t)$, $Var(X_t)$, and $ACF(X_t)$:

$$E(X_t) = E((1+\alpha_1 B + \alpha_2 B^2 + \ldots)Z_t) \\ = E(Z_t) + \alpha_1 E(Z_{t-1}) + \alpha_2 E(Z_{t-2}) + \ldots \\ = 0 + \alpha_1 (0) + \alpha_2 (0) + \ldots = 0$$

$$Var(X_t) = Var((1+\alpha_1 B + \alpha_2 B^2 + \ldots)Z_t) \\ = Var(Z_t) + \alpha_1^2 Var(Z_{t-1}) + \alpha_2^2 Var(Z_{t-2}) + \ldots \\ = \sigma^2 (1+\alpha_1^2 + \alpha_2^2 + \ldots) \\ = \sigma^2 \sum_{i=0}^\infty \alpha_i^2$$
where $\alpha_0 = 1$. Note that a necessary condition for stationarity (constant variance) is that the sum converges ($|\phi| < 1$).

The autocovariance of an MA(q) process, as shown previously, is given by 

$$\gamma(k) = \sigma^2\sum_{i=0}^{q-k}\alpha_i\alpha_{i+k}$$

Then for an AR(1) process, which is the same as an MA($\infty$) process, the autocovariance is given by

$$\gamma(k) = \sigma^2\sum_{i=0}^{\infty}\alpha_i\alpha_{i+k}$$

Then, the autocorrelation is given by

$$\rho(k) = \frac{\gamma(k)}{\gamma(0)} = \frac{\sigma^2\sum_{i=0}^{\infty}\alpha_i\alpha_{i+k}}{\sigma^2\sum_{i=0}^{\infty}\alpha_i\alpha_{i}} = \frac{\sum_{i=0}^{\infty}\alpha_i\alpha_{i+k}}{\sum_{i=0}^{\infty}\alpha_i\alpha_{i}}$$
